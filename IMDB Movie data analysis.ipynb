#### Classify movies based on ratings [Excellent , Good and Average]

```python
data.columns
```

```python
data['Rating']
```

```python
# if (data['Rating']>=7.5):
#     print('Excellent')
# elif (data['Rating']>=4 and data['Rating']<7.5):
#     print('Good')
# else:
#     print('Average')
```

```python
def rating(rating):
    if (rating>=7.0):
        return 'Excellent'
    elif (rating>=6.0 and rating<7.0):
        return 'Good'
    else:
        return 'Average'
```

```python
data['rating_cat']=data['Rating'].apply(rating)
```

```python
data.head()
```

```python
data['rating_cat'].unique()
```

```python
data['rating_cat'].value_counts()
```

```python
sns.countplot(x='rating_cat',data=data)
```

#### Find unique values from genre

```python
data['Genre']
```

```python
data['Genre'].str.split(',')
```

```python
l1=[]
for value in data['Genre']:
    l1.append(value.split(','))
```

```python
l1
```

```python
l2=[]
for item in l1:
    for item1 in item:
        l2.append(item1)
```

```python
l2
```

```python
l3=[]
for item in l2:
    if item not in l3:
        l3.append(item)
```

```python
l3
```

```python
len(l3)
```
