### Plot continuous variables

```python
data.columns
```

```python
cat_val=[]
cont_val=[]
for column in data.columns:
    if data[column].nunique()<=10:
        cat_val.append(column)
    else:
        cont_val.append(column)
```

```python
cat_val
```

```python
cont_val
```

```python
data.hist(cont_val,figsize=(15,6))
plt.show()
```

```python
data.hist(cont_val,figsize=(15,6))
plt.tight_layout()
plt.show()
```
